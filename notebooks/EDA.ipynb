```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('../vehicles_us.csv')
df.head()
```

```python
fig = px.histogram(df, x='odometer')
fig.show()
```

```python
fig = px.scatter(df, x='odometer', y='price')
fig.show()
```

```python
df_clean = df.copy()

# eliminar valores nulos en columnas clave
df_clean = df_clean.dropna(subset=['odometer', 'price'])

# eliminar outliers extremos
df_clean = df_clean[df_clean['odometer'] < 300000]
df_clean = df_clean[df_clean['price'] < 100000]
```

```python
fig = px.scatter(df_clean, x='odometer', y='price')
fig.show()
```
